```python
from textblob import TextBlob
from newspaper import Article

def nlp(text):
  #print(text)

  blob = TextBlob(text)
  sentiment = blob.sentiment.polarity
  print(sentiment)

def nlpFromUrl(url):
  article = Article(url)

  article.download()
  article.parse()
  article.nlp()

  text = article.text

  nlp(text)

def nlpFromText(text):
  nlp(text)
```

```python
nlpFromUrl("https://en.wikipedia.org/wiki/Algebra")
```
